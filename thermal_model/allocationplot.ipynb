```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
allocation_array = [5, 2, 6, 2, 4]
labels = ['Component A', 'Component B', 'Component C', 'Component D', 'Component E']
actual_value = [ 2, 8, 3, 1, 2]

# Create a plot that difference between allocation and actual value
diff_array = np.array(allocation_array) - np.array(actual_value)
x = np.arange(len(labels))
width = 0.5
fig, ax = plt.subplots()
# plot the positive and negative values in different colors
colors = ['red' if x < 0 else 'green' for x in diff_array]
rects1 = ax.bar(x - width/2, diff_array, width, label='Allocation', color=colors)
ax.set_ylabel('Values')
ax.set_title('Allocation vs Actual Value')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
